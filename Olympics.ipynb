```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
data=pd.read_csv('/content/athlete_events.csv')
```

```python
data['Age'].fillna(data['Age'].mean())
data["Height"].fillna(data["Height"].mean())
data["Weight"].fillna(data["Weight"].mean())
```

```python
data.isna().sum()
```

```python
data=data.drop_duplicates()
```

```python
num_data.corr()
```

```python
sns.heatmap(num_data.corr(),annot=True,linecolor='black',linewidths='5')
```

```python
sex_count=data['Sex'].value_counts()
print(sex_count)
```

```python
sns.barplot(x=sex_count.index,y=sex_count.values)
```

```python
plt.figure(figsize=(8 ,4))
t10=data['Sport'].value_counts().head(10)
plt.ylabel('Sport')
plt.xlabel('Count')
plt.title('Top 10 Sports')
sns.barplot(x=t10.values,y=t10.index)
```

```python
t10.plot(kind='pie',autopct='%1.1f%%',shadow=True)
```

```python
plt.figure(figsize=(4,4))
sns.histplot(data['Age'],kde=True,bins=20)
```

```python
plt.figure(figsize=(12 ,8))
sns.scatterplot(x=data['Age'],y=data['Height'],hue=data['Sex'])
```

```python
sns.barplot(y=data['Medal'],x =data['Age'])
```
